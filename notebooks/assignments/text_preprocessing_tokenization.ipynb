```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from konlpy.tag import Okt

english_stopwords = set(stopwords.words('english'))

korean_stopwords = ['은', '는', '이', '가', '을', '를', '에', '에게', '와', '과', '하다', '이다', '되다']

def preprocess_text(text):
    """
    주어진 텍스트를 전처리하는 함수

    인자:
        text (str): 전처리할 텍스트

    반환값:
        list: 전처리된 토큰(단어) 리스트
    """

    text = text.lower()

    text = re.sub(r'http\S+|www.\S+', '', text)
    text = re.sub(r'[^가-힣a-zA-Z\s]', '', text)

    tokens = []
    okt = Okt()

    if any(c for c in text if '가' <= c <= '힣'):
        korean_tokens = okt.nouns(text)
        tokens = [word for word in korean_tokens if word not in korean_stopwords]
    else:
        english_tokens = word_tokenize(text)
        tokens = [word for word in english_tokens if word not in english_stopwords and len(word) > 1]

    return tokens
```

```python

sentences = [
    "This is a test sentence for natural language processing.",
    "자연어 처리는 생각보다 어려운 것 같다.",
    "When I was a young man my father took me in the city.",
    "내일이 월요일이라니 정말 즐거워요!",
    "This website, https://smartlead.hallym.ac.kr/, has good information."
]

processed_sentences = [preprocess_text(sentence) for sentence in sentences]

for i, result in enumerate(processed_sentences):
    print(f"원본 문장 {i+1}: {sentences[i]}")
    print(f"전처리된 토큰: {result}\n")
```
